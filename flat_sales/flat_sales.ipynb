Дополнительно проверим нет ли ошибок в заполнении этажа или этаж больше чем количество этажей в доме.

```python
# <создаем функцию для обобщения категорий этажей>
def floor_short(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == floors_total:
        return 'последний'
    elif floor == 1:
        return 'первый'
    elif floor > floors_total:
        return 'ошибка'
    else:
        return 'другой'
```

Применяем функцию для обобщения категорий этажей ко всей таблице. Посчитаем количество значений в каждой категории.

```python
# <создаем новый столбец floor_short и применяем к нему функцию для обобщения категорий этажей>
df['floor_short'] = df.apply(floor_short, axis = 1)
df['floor_short'].value_counts()
```

### Анализ изменения цены по степени удаленности от центра

Как меняется цена для квартир в Санкт-Петербурге в зависимости от удаления от центра? Логично предположить, что чем дальше к окраинам, тем дешевле. Проверим эту гипотезу.

Для начала сделаем срез по населенным пунктам и оставим только Санкт-Петербург.

```python
# <делаем срез методом query()>
spb = df.query('locality_name == "санкт-петербург" and cityCenters_nearest > 0')
spb.head()
```

Создадим отдельный столбец, в котором посчитаем растояние до центра в километрах и округлим его методом astype(), переведя в цедое число.

```python
spb['to_center'] = spb['cityCenters_nearest'] / 1000
spb['to_center'] = spb['to_center'].astype('int')
spb.head()
```

```python
# <делаем группировку методом groupby() и считаем среднюю цену квадратного метра на определенном километре>
# <переименовываем столбцы>
spb_grouped = spb.groupby('to_center').agg({'price_sq_meter':['median']})
new_names_2 = ['price_sq_meter_spb']
spb_grouped.set_axis(new_names_2, axis = "columns", inplace = True)
spb_grouped
```

```python
# <строим гисторгамму>
spb_grouped.plot(style='o-', grid = True, figsize = (10,3));
```

Мы видим, что самая высокая стоимость квадратного метра в пределах 7 километров от центра, затем она начинает снижаться. Так же есть скачек в 27 километрах, посмотрим что это. Сделаем срез, где укажем расстояние, равное 27 километрам.

```python
display(spb.query('to_center == 27'))
```

Этот всплеск дали 2 квартиры, причем одна из них с очень высокой стоимостью квадратного метра. Соответственно этот всплеск можно игнорировать.

Обозначим центром территорию 0-7 километров. Сделаем срез.

Кажется, что 7 километров - многовато для центра, но именно на этой отметке цена фиксируется и далее начинает падать.

```python
spb_center = spb.query('to_center <= 7')
```
